```python
import pandas as pd
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
df = pd.read_csv("xaks/МС_Республика_Мордовия_Преобразованный.csv")
df = df.drop(["Unnamed: 0"], axis = 1)
df["ЕЛС действующий"] = df["ЕЛС действующий"].apply(bool)
scaler = StandardScaler()
X = np.array(df.drop("Карточка клиента (внешний источник).Индекс финансового риска Значение", axis=1))
X = scaler.fit_transform(X)
y = np.array(df["Карточка клиента (внешний источник).Индекс финансового риска Значение"])
```

```python
history = []
for depth in range(9, 14):
    print(depth)
    for criterion in ("gini", "entropy", "log_loss"):
        for n_estimators in range(50, 500, 50):
            for split in range(2, 5):
                reg = RandomForestClassifier(max_depth = depth, criterion = criterion, n_estimators = n_estimators, min_samples_split = split, n_jobs = -1)
                cvs = cross_val_score(reg, X, y, cv=5, scoring="accuracy")
                mean_cvs = np.mean(cvs)
                history.append((np.abs(mean_cvs), depth, criterion, n_estimators,split))
```
