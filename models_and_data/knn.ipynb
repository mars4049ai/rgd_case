```python
import pandas as pd
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("xaks/МС_Республика_Мордовия_Преобразованный.csv")
```

```python
df = df.drop(["Unnamed: 0"], axis = 1)
```

```python
df["ЕЛС действующий"] = df["ЕЛС действующий"].apply(bool)
```

```python
df.info()
```

```python
df.sort_index(ascending=True)
df["Карточка клиента (внешний источник).Индекс финансового риска Значение"].unique()
```

```python
scaler = StandardScaler()
X = np.array(df.drop("Карточка клиента (внешний источник).Индекс финансового риска Значение", axis=1))
X = scaler.fit_transform(X)
y = np.array(df["Карточка клиента (внешний источник).Индекс финансового риска Значение"])
```

```python
y
```

```python
table = df.corr()
table
plt.figure(figsize=(10, 8))
sns.heatmap(table, annot=True, linewidth=0.5)
plt.show()
```

```python
history = []
for k in range(1, 30):
    print(k)
    reg = KNeighborsClassifier(n_neighbors = k)
    cvs = cross_val_score(reg, X, y, cv=5, scoring="accuracy")
    mean_cvs = np.mean(cvs)
    history.append((np.abs(mean_cvs), k))
```
