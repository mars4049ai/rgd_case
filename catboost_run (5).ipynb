```python
import pandas as pd
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from catboost import CatBoostClassifier

df = pd.read_csv("РЖД train/Выгрузка_маркетинговые списки/МС_Республика Татарстан_Преобразованный.csv")
df = df.drop(["Unnamed: 0"], axis = 1)
df["ЕЛС действующий"] = df["ЕЛС действующий"].apply(bool)
scaler = StandardScaler()
X = np.array(df.drop("Карточка клиента (внешний источник).Индекс финансового риска Значение", axis=1))
X = scaler.fit_transform(X)
y = np.array(df["Карточка клиента (внешний источник).Индекс финансового риска Значение"])

history = []
for depth in range(8, 15):
    for lr in np.arange(0.05, 0.5, 0.05):
        for iterr in range(100, 900, 50):
            reg = CatBoostClassifier(iterations = iterr, learning_rate = lr, depth = depth, task_type="GPU")
            cvs = cross_val_score(reg, X, y, cv=5, scoring="accuracy")
            mean_cvs = np.mean(cvs)
            history.append((np.abs(mean_cvs),depth, lr,iterr))
```
